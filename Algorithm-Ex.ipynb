```python
def count_char_frequencies(s):
    freq_dict = {}
    for char in s:
        if char in freq_dict:
            freq_dict[char] += 1
        else:
            freq_dict[char] = 1
    return freq_dict
```

```python
def squares(n):
    return [i * i for i in range(n)]
```
